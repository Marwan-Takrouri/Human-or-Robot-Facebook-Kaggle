```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

train_df = pd.read_csv("./Resources/train.csv")
test_df = pd.read_csv("./Resources/test.csv")
bids_df = pd.read_csv("./Resources/bids.csv")
```

```python
# Groupby auction
auctions = pd.DataFrame(data = bids_df['auction'].unique(), columns = ['auction'],
                    index = bids_df['auction'].unique())

text_cols = ['bidder_id', 'merchandise', 'device', 'country', 'ip', 'url']
for var in text_cols:
    bids_df[var] = var + "_" + bids_df[var].fillna("")
    count_str = var + '_nunique_num'
    auctions[count_str]  = bids_df.groupby('auction')[var].nunique()

auctions.head()
```

```python
# Groupby bidders
bids = pd.DataFrame(data = bids_df['bidder_id'].unique(), columns = ['bidder_id'],
                    index = bids_df['bidder_id'].unique())

text_cols = ['auction', 'merchandise', 'device', 'country', 'ip', 'url']
for var in text_cols:
    bids_df[var] = var + "_" + bids_df[var].fillna("")
    count_str = var + '_nunique_num'
    bids[count_str]  = bids_df.groupby('bidder_id')[var].nunique()

bids.head()
```
