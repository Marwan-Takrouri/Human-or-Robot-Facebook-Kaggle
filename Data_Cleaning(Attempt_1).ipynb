```python
# Merge on train df
dataset = pd.merge(train,bids, on = "bidder_id", how ="left")
dataset.to_csv("dataset.csv")
```

```python
# Reload the dataset 
df = pd.read_csv("dataset.csv")
df.head()
```

```python
len(df.index)
```

```python
df.set_index("Unnamed: 0", inplace= True)
```

```python
df.head()
```

```python
df.loc[df["outcome"] == 1.0]
```

```python
# Find number of null values 
df.isnull().sum()
```

```python
df[df["bid_id"].isnull()]
```

```python
# All of the bid_id null values are by humans - same for columns other than country..... so they can be removed 
# (https://stackoverflow.com/questions/13413590/how-to-drop-rows-of-pandas-dataframe-whose-value-in-a-certain-column-is-nan)
df = df[df["bid_id"].notna()]
```

```python
df.isnull().sum()
```

```python
# Rows with null country values
country_null = df[df["country"].isnull()]
```

```python
country_null
```

```python
# Some robot bids observed in the country_null dataframe.... let's see what the total number of rows with 
# outcome = 1 and country = null is 
country_null = pd.DataFrame(country_null)
```

```python
country_null[country_null["outcome"] == 1.0]
```

```python
# Very small minority of robotic bids have null for country.... we will remove those 
df = df[df["country"].notna()]
```

```python
df.isnull().sum()
```

```python
df.to_csv("cleaned_dataset.csv", index = False)
```
